```python
import requests
from bs4 import BeautifulSoup
!pip install xlsxwriter
import pandas as pd

URL = "https://ev-database.org/car/1991/Tesla-Model-3"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find(id="main-data")
```

```python
print(results.prettify())
```

```python
#job_elements = results.find_all("div", class_="data-table has-legend")
job_elements = results.find_all("div", class_="inline-block")
```

```python
for job_element in job_elements:
    print(job_element, end="\n"*2)
```

```python
for job_element in job_elements:
    #title_element = job_element.find("h2")
    data_element = job_element.find("table")
    #company_element = job_element.find("h3", class_="company")
    #location_element = job_element.find("p", class_="location")
    #print(title_element.text.strip())
    print(data_element.text.strip())
    #print(company_element.text.strip())
    #print(location_element.text.strip())
    print()
```

```python
result = data_element.text.strip()
print(result)
```

```python
df = pd.DataFrame(result)
df.to_excel("final.xlsx")
```
